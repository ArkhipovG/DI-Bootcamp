```python

# Assuming df is your DataFrame and it has been read in previously
# Convert 'Order Date' to datetime if it isn't already

# Extract year from 'Order Date'
df['Order Year'] = df['Order Date'].dt.year

# Group by 'Order Year' and sum 'Sales'
df_grouped = df.groupby('Order Year')['Sales'].sum().reset_index()

# Create a line chart
fig = go.Figure()

# Add line trace to figure with aggregated data
fig.add_trace(go.Scatter(x=df_grouped['Order Year'], y=df_grouped['Sales'], mode='lines+markers', name='Sales'))

# Customize the layout
fig.update_layout(title='Sales Trends Over the Years',
                  xaxis_title='Order Year',
                  yaxis_title='Sales',
                  template='plotly_dark')

# Show the figure
fig.show()
```

```python
# Group by 'State' and sum 'Sales'
df_grouped = df.groupby('State_abbrev')['Sales'].sum().reset_index()

# Create the choropleth map with aggregated data
fig = px.choropleth(df_grouped,
                    locations='State_abbrev',
                    locationmode='USA-states',
                    color='Sales',
                    color_continuous_scale=px.colors.sequential.Plasma,
                    title='Sales Distribution by States')

# Customize layout if needed
fig.update_layout(geo_scope='usa')  # Focus on USA

# Show the figure
fig.show()
```
