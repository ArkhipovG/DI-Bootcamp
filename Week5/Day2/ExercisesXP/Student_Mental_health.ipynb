```python
import pandas as pd
df = pd.read_csv('/content/sample_data/Student Mental health.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
# Drop rows with missing data in column: 'Age'
df = df.dropna(subset=['Age'])
df.isnull().sum()
```

```python
# Rename column 'Choose your gender' to 'Gender'
df = df.rename(columns={'Choose your gender': 'Gender'})
# Rename column 'What is your course?' to 'Course'
df = df.rename(columns={'What is your course?': 'Course'})
# Rename column 'Your current year of Study' to 'Study year'
df = df.rename(columns={'Your current year of Study': 'Study year'})
# Rename column 'What is your CGPA?' to 'CGPA'
df = df.rename(columns={'What is your CGPA?': 'CGPA'})
# Rename column 'Do you have Depression?' to 'Depression'
df = df.rename(columns={'Do you have Depression?': 'Depression'})
# Rename column 'Do you have Anxiety?' to 'Anxiety'
df = df.rename(columns={'Do you have Anxiety?': 'Anxiety'})
# Rename column 'Do you have Panic attack?' to 'Panic attack'
df = df.rename(columns={'Do you have Panic attack?': 'Panic attack'})
# Rename column 'Did you seek any specialist for a treatment?' to 'Treatment'
df = df.rename(columns={'Did you seek any specialist for a treatment?': 'Treatment'})
```

```python
df
```

```python
df['CGPA'].unique()
```

```python
# Replace all instances of "3.50 - 4.00 " with "3.50 - 4.00" in column: 'What is your CGPA?'
df['CGPA'] = df['CGPA'].str.replace("3.50 - 4.00 ", "3.50 - 4.00", case=False, regex=False)
```

```python
# prompt: Use Seaborn’s histplot to create a histogram of the CGPA categories.

import seaborn as sns
import matplotlib.pyplot as plt

# Create a histogram of the CGPA categories
sns.histplot(df['CGPA'].sort_values(), bins=6)

# Add title and labels
plt.title('Histogram of CGPA Categories')
plt.xlabel('CGPA')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
# Group the data by gender and anxiety
grouped_data = df.groupby(['Gender', 'Anxiety']).size().unstack()

# Create a bar plot
grouped_data.plot(kind='bar', stacked=True)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add title and labels
plt.title('Proportion of Students Experiencing Anxiety by Gender')
plt.xlabel('Gender')
plt.ylabel('Proportion')

# Show the plot
plt.show()
```

```python
# Convert panic attack responses to numeric
df['Panic attack'] = df['Panic attack'].replace({'Yes': 1, 'No': 0})

# Create a scatterplot with age on the x-axis and panic attacks on the y-axis
sns.scatterplot(x='Age', y='Panic attack', data=df, hue='Gender')

# Add title and labels
plt.title('Scatterplot of Age vs. Panic Attacks')
plt.xlabel('Age')
plt.ylabel('Panic Attacks (Yes=1, No=0)')

# Show the plot
plt.show()
```

```python
# Select relevant columns
subset_df = df[['Age', 'CGPA', 'Depression', 'Anxiety', 'Panic attack']]
subset_df['Depression'] = subset_df['Depression'].replace({'Yes': 1, 'No': 0})
subset_df['Anxiety'] = subset_df['Anxiety'].replace({'Yes': 1, 'No': 0})

# Create a pairplot with hue based on Depression
sns.pairplot(subset_df, hue='CGPA')

# Show the plot
plt.show()
```

```python
# Calculate the correlation matrix for the selected subset of the dataset
correlation_matrix = subset_df.corr()

# Create a heatmap using Seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

# Add title and labels
plt.title('Correlation Matrix for Selected Subset')

# Show the plot
plt.show()
```
