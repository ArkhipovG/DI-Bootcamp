```python
import pandas as pd
df2019 = pd.read_csv('2019.csv')
```

```python
df2019
```

```python
import matplotlib.pyplot as plt
df2019.plot.scatter(x='Social support', y='Score')
plt.show()
```

```python
import random as rd
countries = []
for _ in range(7):
  countries.append(rd.choice(df2019['Country or region']))
df_rc = df2019[df2019['Country or region'].isin(countries)]
# print(df_rc)

fig, axes = plt.subplots(2, 1, figsize = (7, 6), sharex=True)
df_rc.plot(x = 'Country or region', y = 'GDP per capita', kind = 'bar', ax = axes[0])
df_rc.plot(x = 'Country or region', y = 'Healthy life expectancy', kind = 'bar', ax = axes [1])
plt.show()
```

```python
fig, axes = plt.subplots(figsize = (7, 6))
axes.bar(df_rc['Country or region'],df_rc['GDP per capita'])
axes.plot(df_rc['Country or region'],df_rc['Healthy life expectancy'], marker = 'o', color ='red')
plt.show()
```
