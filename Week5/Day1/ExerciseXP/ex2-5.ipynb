```python
import pandas as pd
df = pd.read_csv('/content/sample_data/sales_data.csv')
```

```python
df
```

```python
total_quantity = df['quantity'].sum()
print(f"Total quantity of products sold: {total_quantity}")
```

```python
highest_revenue_category = df.groupby('category')['revenue'].sum().sort_values(ascending=False).index[0]
highest_revenue = df.groupby('category')['revenue'].sum().sort_values(ascending=False)[0]
print(f"Category with highest revenue: {highest_revenue_category}")
print(f"Revenue generated by the category: ${highest_revenue:.2f}")
```

```python
average_revenue = df['revenue'].sum() / df['quantity'].sum()
print(f"Average revenue per sale: ${average_revenue:.2f}")
```

```python
df['quater'] = pd.to_datetime(df['date']).dt.quarter
```

```python
# Group data by quarter and calculate total revenue
quarterly_revenue = df.groupby('quater')['revenue'].sum()

# Create a bar chart
quaters = ['Q1', 'Q2', 'Q3', 'Q4']
plt.bar(quaters, quarterly_revenue)

# Set labels and title
plt.xlabel("Quarter")
plt.ylabel("Total Revenue")
plt.title("Total Revenue by Quarter")

# Show the plot
plt.show()
```

```python
# Generate x values from 0 to 10
x_values = list(range(-10, 11))

# Generate y values for y = x^2
y_values = [x**2 for x in x_values]

# Plot
plt.plot(x_values, y_values, marker='o', linestyle='-')

# Add labels and title
plt.title('Line Plot of y = x^2')
plt.xlabel('x')
plt.ylabel('y')

# Show the plot
plt.grid(True)
plt.show()
```

```python
# Define product names and sales values
products = ['A', 'B', 'C', 'D']
sales_values = [15, 30, 45, 20]

# Create a bar chart
plt.bar(products, sales_values)

# Add labels and title
plt.xlabel("Products")
plt.ylabel("Sales Values")
plt.title("Sales Data for Different Products")

# Show the plot
plt.show()
```

```python
# Define data values for each fruit
fruit = ['Apples', 'Bananas', 'Cherries', 'Dates']
quantity = [40, 30, 20, 10]

# Define colors for each fruit
colors = ['red', 'yellow', 'pink', 'brown']

# Create a pie chart
plt.pie(quantity, labels=fruit, colors=colors, autopct='%1.1f%%')

# Add a legend
plt.legend(loc='upper left')

# Add a title
plt.title("Favorite Fruits")

# Show the plot
plt.show()
```
