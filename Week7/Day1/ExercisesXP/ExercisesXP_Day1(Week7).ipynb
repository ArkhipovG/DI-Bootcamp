```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
url = 'https://www.sportsworld.com/'
response = urlopen(url)
html_sportsworld = response.read()
soup_sportsworld = BeautifulSoup(html_sportsworld, 'html.parser')
print(soup_sportsworld.prettify())
```

```python
title = soup_sportsworld.title.get_text()
title
```

```python
paragraphs = soup_sportsworld.find_all('p')

for paragraph in paragraphs:
    print(paragraph.get_text())
```

```python
links = soup_sportsworld.find_all('a')
hrefs = []
for link in links:
    href = link.get('href')
    if href:
        hrefs.append(href)

for href in hrefs:
  print(href)
```

## Exercise 4

```python
def _check_is_title(url):
  response = requests.get(url)
  soup = BeautifulSoup(response.text, 'html.parser')
  title = soup.title.get_text()
  if not title:
    print('There is no title')
  else:
    print(f'Title is: {title}')
```

```python
_check_is_title('https://en.wikipedia.org/wiki/Main_Page/')
```

## Exercise 6

```python
url = 'https://www.imdb.com/list/ls091294718/'
response = requests.get(url)
movies_soup = BeautifulSoup(response.text, 'html.parser')
```

```python
names = []
years = []
directors = []
stars = []

for movie in movies_soup.find_all('div', class_='lister-item mode-detail'):
    movie_name = movie.h3.a.text
    movie_year = movie.h3.find('span', class_='lister-item-year text-muted unbold').text
    movie_summary = movie.find_all('p', class_='text-muted')[1].text

    parts = movie_summary.split('|')
    movie_year = movie_year.replace("(", "").replace(")", "")
    movie_directors = parts[0].strip().split(':')[1].split(',')
    movie_directors = [director.strip().replace('\n', '') for director in movie_directors]
    movie_stars = parts[1].strip().split(':')[1].split(',')
    movie_stars = [star.strip().replace('\n', '') for star in movie_stars]
    movie_stars_str = ', '.join(movie_stars)
    movie_director_str = ', '.join(movie_directors)
    names.append(movie_name)
    years.append(movie_year)
    directors.append(movie_director_str)
    stars.append(movie_stars_str)

```

```python
import pandas as pd

data = pd.DataFrame({
  "name": names,
  "year": years,
  "director": directors,
  'stars': stars
})
```
