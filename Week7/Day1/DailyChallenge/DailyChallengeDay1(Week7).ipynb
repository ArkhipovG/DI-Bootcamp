```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://github.com/topics'
response = requests.get(url)
github_soup = BeautifulSoup(response.text, 'html.parser')
```

```python
response.status_code
```

```python
print(github_soup.prettify()[:100])
```

```python
# Save the HTML content to a file named webpage.html. Ensure you handle the text encoding correctly.

with open('webpage.html', 'w', encoding='utf-8') as file:
  file.write(github_soup.prettify())
```

```python
with open('webpage.html', 'r', encoding='utf-8') as file:
  html_content = file.read()

soup = BeautifulSoup(html_content, 'html.parser')


title = soup.title.get_text().strip()
title
```

```python
# prompt: Identify two distinct pieces of information on the webpage to extract (e.g., titles of the topics and their descriptions)

# Extract titles of the topics
topic_titles = [topic.text.strip() for topic in soup.find_all( class_='f3 lh-condensed mb-0 mt-1 Link--primary')]

# Extract descriptions of the topics
topic_descriptions = [description.text.strip() for description in soup.find_all(class_='f5 color-fg-muted mb-0 mt-1')]

print(f"The length of topic_titles is: {len(topic_titles)}")
print(f"The length of topic_descriptions is: {len(topic_descriptions)}")
```

```python
github_topics = pd.DataFrame({
  "title": topic_titles,
  "description": topic_descriptions,
})
```
