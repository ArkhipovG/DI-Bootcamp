```python
import pandas as pd
mental_health_data = pd.read_csv('survey.csv')
mental_health_data
```

```python
cleaned_mental_health_data = mental_health_data
```

```python
cleaned_mental_health_data.info()
```

```python
# Drop column 'comments' because its 87% missing values so its not relevant for us
cleaned_mental_health_data = cleaned_mental_health_data.drop(columns=['comments'])
```

```python
# Drop column 'comments' because  its not relevant for us
cleaned_mental_health_data = cleaned_mental_health_data.drop(columns=['state'])
```

```python
cleaned_mental_health_data
```

```python
cleaned_mental_health_data['Gender'].replace(['man', 'Male ', 'm', 'maile', 'Male (CIS)','Cis Man' ,'Cis Male', 'msle', 'Mail', 'Make', 'Mal', 'Malr', 'Male-ish', 'Man', 'M', "male"], 'Male', inplace=True)
cleaned_mental_health_data['Gender'].replace([ 'Female ', 'Female (cis)','cis-female/femme', 'feMale', 'F', 'f', 'woman', 'Woman', 'Cis Female', 'Cis FeMale', 'femail', 'Femake', 'female'], 'Female', inplace=True)
cleaned_mental_health_data.loc[~cleaned_mental_health_data['Gender'].isin(['Male', 'Female']), 'Gender'] = 'Other'
```

```python
cleaned_mental_health_data
```

```python
Q1A = cleaned_mental_health_data['Age'].quantile(0.25)
Q3A = cleaned_mental_health_data['Age'].quantile(0.75)
IQRA = Q3A - Q1A
lower_boundA = Q1A - 1.5 * IQRA
upper_boundA = Q3A + 1.5 * IQRA
print(lower_boundA,upper_boundA)
outliers_data_ages = cleaned_mental_health_data[(cleaned_mental_health_data['Age'] < lower_boundA) | (cleaned_mental_health_data['Age'] > upper_boundA)]
new_cleaned_mental_data = cleaned_mental_health_data[(cleaned_mental_health_data['Age'] >= lower_boundA) & (cleaned_mental_health_data['Age'] <= upper_boundA)]
```

```python
outliers_data_ages
```

```python
new_cleaned_mental_data
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
new_cleaned_mental_data.loc[:, 'Age_normalized'] = scaler.fit_transform(new_cleaned_mental_data[['Age']])
```

```python
new_cleaned_mental_data
```

```python
# Drop rows with missing data in columns: 'self_employed', 'work_interfere'
new_cleaned_mental_data = new_cleaned_mental_data.dropna(subset=['self_employed', 'work_interfere'])
```

```python
new_cleaned_mental_data
```

```python
test_data = new_cleaned_mental_data
# Define age groups
age_bins = [18, 25, 35, 45, 50]  # Define your age groups as needed
age_labels = ['18-24', '25-34', '35-44', '45-49']
test_data.loc[:,'Age_Group'] = pd.cut(test_data['Age'], bins=age_bins, labels=age_labels, right=False)

# Group the data by age group and calculate the proportion of respondents who sought treatment
age_group_distribution = test_data.groupby('Age_Group')['treatment'].value_counts(normalize=True).unstack() * 100
```

```python
age_group_distribution
```

```python
gender_distribution = test_data.groupby('Gender')['treatment'].value_counts(normalize=True).unstack() * 100
gender_distribution
```

```python
country_distribution2 = test_data.groupby('Country')['treatment'].value_counts(normalize=True).unstack() * 100
country_distribution2
```

```python
respondent_counts = test_data.groupby('Country')['treatment'].count()

min = 10
filtered_countries = respondent_counts[respondent_counts>= min].index

filtered_menthal_health_rates = test_data[test_data['Country'].isin(filtered_countries)].groupby('Country')['treatment'].apply(lambda x: (x == "Yes").mean()).reset_index()
sorted_filtered_rates_desc = filtered_menthal_health_rates.sort_values('treatment', ascending= False)
sorted_filtered_rates_asc = filtered_menthal_health_rates.sort_values('treatment', ascending= True)

print("Countries with the highest reported rates of mental health issues (filtered):")
sorted_filtered_rates_desc.head()
```
