```python
chemical_elements = final_df.columns[1:-1]
for element in chemical_elements:
    plt.figure(figsize=(8, 6))
    plt.scatter(final_df[element], final_df['quality'], alpha=0.5)
    plt.title(f'Scatter plot: {element} vs Quality')
    plt.xlabel(element)
    plt.ylabel('Quality')
    plt.grid(True)
    plt.show()
```
