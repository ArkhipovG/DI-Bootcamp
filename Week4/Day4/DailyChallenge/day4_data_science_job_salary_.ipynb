```python
import pandas as pd
df = pd.read_csv('datascience_salaries.csv')
```

```python
min_salary = df['salary'].min()
max_salary = df['salary'].max()

# Apply Min-Max normalization to the 'salary' column
df['normalized_salary'] = (df['salary'] - min_salary) / (max_salary - min_salary)
```

```python
df
```

```python
final_df = df.drop(columns='Unnamed: 0')
```

```python
final_df
```

```python
final_df['experience_level'] = pd.Categorical(final_df['experience_level'], categories=['Entry', 'Mid', 'Senior', 'Executive'], ordered=True)
```

```python
final_df.groupby('experience_level').describe()
```

```python
grouped_data = final_df.groupby('experience_level')['salary'].agg(['mean', 'median'])
```
