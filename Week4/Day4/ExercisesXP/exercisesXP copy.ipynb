```python
import pandas as pd
train_data = pd.read_csv('/Users/grigorijarhipov/Documents/Developers.Institute/GITHUB/DI-Bootcamp/Week4/Day4/ExercisesXP/titanic/train.csv')
train_data
```

```python
print(f" train_data duplicates: {train_data.duplicated().sum()}")
```

```python
print(f" train_data missing values count: \n{train_data.isnull().sum()}\n")
```

```python
# Drop column: 'Cabin' beacause there is 77% missing values
cleaned_train_data = train_data.drop(columns=['Cabin'])
cleaned_train_data
```

```python
# Replace gaps forward from the previous valid value in: 'Age'
cleaned_train_data = cleaned_train_data.fillna({'Age': cleaned_train_data['Age'].ffill()})
cleaned_train_data
```

```python
# Replace missing values with the most common value of each column in: 'Embarked'
cleaned_train_data = cleaned_train_data.fillna({'Embarked': cleaned_train_data['Embarked'].mode()[0]})
cleaned_train_data
```

```python
cleaned_train_data['Family Size'] = cleaned_train_data['SibSp'] + cleaned_train_data['Parch']
cleaned_train_data
```

```python
cleaned_train_data["Title"] = cleaned_train_data['Name'].str.split(',').str[0]
cleaned_train_data
```

```python
from sklearn.preprocessing import OneHotEncoder

# Create LabelEncoder object

onehot_encoder = OneHotEncoder()
encoded_cleaned_train_data = cleaned_train_data
# Encode the categorical variable
encoded = onehot_encoder.fit_transform(encoded_cleaned_train_data['Sex'].to_numpy().reshape(-1, 1)).toarray()
encoded_cleaned_train_data['Sex']= encoded
```
